```python
%pylab inline
import csv

analfabetismo = []
inasistencia_escolar = []
alta_depend_econ = []
sin_fuente_agua = []

n_municipios = 0

with open('pobreza.csv', 'rb') as csvfile:
    reader = csv.reader(csvfile, delimiter=' ', quotechar='"')
    for row in reader:
        analfabetismo.append(row[11])
        inasistencia_escolar.append(row[13])
        alta_depend_econ.append(row[21])
        sin_fuente_agua.append(row[29])
        n_municipios += 1
        
        
n_municipios = n_municipios - 1
n_factores = 4
pobreza= empty((n_municipios,n_factores))

for i in range(n_municipios):
    pobreza[i][0] = analfabetismo[i+1]
    pobreza[i][1] = inasistencia_escolar[i+1]
    pobreza[i][2] = alta_depend_econ[i+1]
    pobreza[i][3] = sin_fuente_agua[i+1]
    

promedio = zeros(n_factores)

for i in range(n_factores):
    promedio[i] = mean(pobreza[i,:])
    
covarianzas = zeros((n_factores,n_factores))

for i in range(n_factores):
    for j in range(n_factores):
        covarianzas[i,j] =  (sum((pobreza[:,i] - promedio[i]) * (pobreza[:,j] - promedio[j])))/(n_municipios-1.0)

print "la matriz de covarianzas es:"            
print covarianzas


imshow(log10(abs(covarianzas))+1E-30)
```

```python
eigenvalues = -sort(-linalg.eig(covarianzas)[0])

traza= np.sum(eigenvalues)

n = zeros(n_factores)
n[0]=eigenvalues[0]/traza
No_ValoresPropiosImportantes = 0
for i in range(1,n_factores):
    n[i] = n[i-1] + eigenvalues[i]/traza
    
    if (n[i]>0.9):
        No_ValoresPropiosImportantes = i+1
        break
        
print "Los valores propios son:"
print eigenvalues
print "El número de valores propios importantes es:"        
print No_ValoresPropiosImportantes
 
eigen_importantes = zeros(No_ValoresPropiosImportantes)
for i in range(No_ValoresPropiosImportantes):
    eigen_importantes[i] = eigenvalues [i]
    
print "Los valores propios importantes son:"
print eigen_importantes
```
